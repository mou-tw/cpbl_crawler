```python
import pandas as pd
import os
```

```python
ht_lst  = os.listdir('tmp_csv/')
```

```python
hit_lst_atk = [i for i in ht_lst if i.split('_')[1]=='atk.csv']
hit_lst_dfs = [i for i in ht_lst if i.split('_')[1]=='dfs.csv']
```

```python
print(len(hit_lst_atk))
print(len(hit_lst_dfs))
```

```python
%%time
atk_agg_lst ,dfs_agg_lst = [], []
for i in hit_lst_atk:
    tmp_df = pd.read_csv(f'tmp_csv/{i}')
    atk_agg_lst.append(tmp_df)
for i in hit_lst_dfs:
    tmp_df = pd.read_csv(f'tmp_csv/{i}')
    dfs_agg_lst.append(tmp_df)    
```

```python
atk_agg_df = pd.concat(atk_agg_lst)
dfs_agg_df = pd.concat(dfs_agg_lst)
```
